```python
import matplotlib.pyplot as plt
import pandas as pd

# Model performance metrics
results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Gaussian Naive Bayes",
        "SVM",
        "KNN"
    ],
    "Accuracy": [0.620, 0.381, 0.693, 0.943],
    "Precision": [0.100, 0.072, 0.121, 0.167],
    "Recall": [0.84, 0.98, 0.84, 0.04],
    "F1-score": [0.178, 0.134, 0.211, 0.065]
})

results.set_index("Model").plot(kind="bar", figsize=(10,6))

plt.title("Comparison of Machine Learning Model Performance")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=15)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()

plt.savefig("images/model_comparison.png", dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Number of patients without and with stroke
labels = ['No Stroke', 'Stroke']
counts = [4861, 249]

plt.figure(figsize=(6,5))
plt.bar(labels, counts)

plt.title("Distribution of Stroke Cases")
plt.xlabel("Stroke Status")
plt.ylabel("Number of Patients")

for i, value in enumerate(counts):
    plt.text(i, value + 50, str(value), ha='center', fontsize=11)

plt.tight_layout()
plt.savefig("images/class_distribution.png", dpi=300)
plt.show()
```
